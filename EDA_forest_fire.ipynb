```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as pyplot
%matplotlib inline
```

```python
dataset = pd.read_csv('Algerian_forest_fires_dataset.csv')
dataset.head()
```

```python
dataset.info()
```

#### data cleaning

```python
## missing values

dataset[dataset.isnull().any(axis=1)]
```

```python
# since the dataset has two regions dataset is converted into two sets based on region from 122th index, so we are gonna make new colomn based on the region
#1: "Bejaia Region Dataset" - 0
#2: "Sidi-Bel Abbes Region Dataset" - 1

# add new column with region
```

```python
dataset.loc[:122,"Region"] = 0
dataset.loc[122:,"Region"] = 1
df = dataset
```

```python
df.info()
```

```python
df[['Region']] = df[['Region']].astype(int)
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
## remove the null values

df = df.dropna().reset_index(drop=True)
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
df.iloc[[122]]
```

```python
df = df.drop(122).reset_index(drop=True)
```

```python
df.iloc[[122]]
```

```python
df.columns
```

```python
## fix spaces in column names

df.columns = df.columns.str.strip()
df.columns
```

```python
## change the required columns as integer data type

df[['month','day','year','Temperature','RH','Ws']] = df[['month','day','year','Temperature','RH','Ws']].astype(int)
```

```python
df.info()
```

```python
### changing the other columns to float data type

objects = [features for features in df.columns if df[features].dtypes == 'O']
```

```python
for i in objects:
    if i!='Classes':
        df[i]=df[i].astype(float)
```

```python
df.info()
```

```python
objects
```

```python
df.describe()
```

```python
df.head()
```

```python
## lets save the cleaned dataset

df.to_csv('Algerian_forest_fires_cleaned_dataset.csv',index=False)
```

### exploratory data analysis

```python
# drop day month and year
df_copy = df.drop(['day','month','year'],axis=1)
df_copy.head()

df_copy['Classes'].value_counts()
```

```python
## encoding of the categories in classes
df_copy['Classes'] = np.where(df_copy['Classes'].str.contains('not fire'),0,1)
df_copy.head()
```

```python
##categories in classes

df_copy['Classes'].value_counts()
```

```python
## plot density plot for all features
from matplotlib import pyplot as plt
plt.style.use("ggplot")
df_copy.hist(bins=50,figsize=(20,15))
plt.show()
```

```python
#percentage for pie chart
percentage = df_copy['Classes'].value_counts(normalize=True)*100
```

```python
## plotting piechart
classlabels=['Fire','Not fire']
plt.figure(figsize=(12,7))
plt.pie(percentage,labels=classlabels,autopct='%1.1f%%')
plt.show()
```
